```python
# STEP 1: IMPORT LIBRARIES
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/AB_NYC_2019.csv')  # Use this in Google Colab
df.head()
```

```python
# Dataset summary
df.info()

# Check for unique columns, column types, and possible issues
print("Unique columns:", df.columns)
print("\nSummary statistics:\n", df.describe(include='all'))
```

```python
# Missing values per column
df.isnull().sum()
```

```python
# Fill numerical missing value
df['reviews_per_month'].fillna(df['reviews_per_month'].median(), inplace=True)

# Optionally drop rows with crucial missing values
df.dropna(subset=['name', 'host_name'], inplace=True)

# Recheck missing values
df.isnull().sum()
```

```python
# Check for duplicates
duplicates = df.duplicated()
print(f"Total Duplicates: {duplicates.sum()}")

# Drop duplicates
df = df.drop_duplicates()
```

```python
# Standardize column case
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]

# Check and standardize string columns (e.g., neighbourhood_group)
df['neighbourhood_group'] = df['neighbourhood_group'].str.title().str.strip()
df['room_type'] = df['room_type'].str.title().str.strip()

# Check data types
df.dtypes
```

```python
# Remove outliers using IQR
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Filter data
filtered_df = df[(df['price'] >= Q1 - 1.5 * IQR) & (df['price'] <= Q3 + 1.5 * IQR)]
print(f"Original size: {df.shape}, After outlier removal: {filtered_df.shape}")
```

```python
# Save cleaned dataset
filtered_df.to_csv('/content/AB_NYC_2019_cleaned.csv', index=False)
```
